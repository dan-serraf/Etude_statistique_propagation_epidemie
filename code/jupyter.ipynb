### Question 2

Dans le modèle SIR, pour une population donnée,
on étudie la taille de trois sous-populations au cours du temps : <br>
- S(t) les personnes Saines au temps t
- I(t) les personnes Infectées au temps t
- R(t) les personnes Retirées au temps t <br>

Les personnes saines n’ont pas encore été touchées par le virus,
alors que les personnes retirées sont guéries, et donc immunisées.
Les personnes retirées ne sont donc plus prises en compte.
Donc le modèle SIR ne s’occupe pas directement de prédire la mortalité de l’épidémie. <br>

De plus, ce modèle ne prend pas en compte l'âge des personnes et l'environnement.
La propagation peut différer selon les pays ou les villes.
Et le virus peut se comporter d'une manière différente selon l'âge de la personne.
Rajouter ces paramètres serait donc une bonne amélioration à faire.
